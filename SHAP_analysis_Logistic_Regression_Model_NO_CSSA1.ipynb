```python
from  matplotlib.colors import LinearSegmentedColormap
cmap=LinearSegmentedColormap.from_list('rg',["green", "orange"], N=256) 
```

```python
## Import best model (KNN)
log_reg = joblib.load("../top_models/best_logistic_regression.pkl")
```

```python
## Import dataframe for 70_30 feature selected data.
name = "../data/feature_selected_train_dataset_70_30.csv"
df = pd.read_csv(name)

## Separate into features and outcomes. Remove the CSSA1 variable.
X_train = df.drop(['outcome', "CSSA Score Week 1"], axis=1, inplace=False)
y_train = df["outcome"]

## Scale X with standard scaler for models that require it.
X_train_scaled = StandardScaler().fit_transform(X_train)
```

```python
## Import dataframe for 70_30 feature selected data.
name = "../data/feature_selected_test_dataset_70_30.csv"
df = pd.read_csv(name)

## Separate into features and outcomes. Remove the CSSA1 variable.
X_test = df.drop(['outcome', "CSSA Score Week 1"], axis=1, inplace=False)
y_test = df["outcome"]

## Scale X with standard scaler for models that require it.
X_test_scaled = StandardScaler().fit_transform(X_test)
```

```python
## Refit model to features lacking CSSA1
log_reg = log_reg.fit(X_train_scaled, y_train)
```

```python
## Create list of feature names
feature_names = X_test.columns
```

```python
## Create dataframe with scaled data and columns with feature names for test data
X_frame_test = pd.DataFrame(data=X_test_scaled, columns=feature_names)
```

```python
## Create explainer and shap values for model with 
shap.initjs()
explainer = shap.KernelExplainer(log_reg._final_estimator.predict, X_frame_test)
shap_values = explainer.shap_values(X_frame_test)
```

```python
## Create shap summary plot
shap.summary_plot(shap_values, X_frame_test, show=False,  cmap=cmap)
plt.savefig('../figures/log_reg_SHAP_summary_plot_NO_CSSA1.png')
```

```python
## Create shap summary plot as standard feature importance plot (bar plot)
shap.summary_plot(shap_values, X_frame_test, show=False, plot_type='bar')
plt.savefig('../figures/log_reg_SHAP_bar_summary_plot_NO_CSSA1.png')
```
